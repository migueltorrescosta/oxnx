```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
#ex 2

def findGCSandMeanC (n, p):
    G = nx.erdos_renyi_graph (n, p)
    return np.mean (nx.degree(G).values()), 1.0 * (len ( sorted(nx.connected_components(G), key=len, reverse=True) [0])  - 1) / n

n = 200
repeatNum = 3
vals = []

for p in np.linspace (0, 0.1, 100):
    C = 0.0
    S = 0.0
    for r in range (repeatNum):
        c,s = findGCSandMeanC (n, p)
        C += c
        S += s
    C /= repeatNum
    S /= repeatNum
    if C < 4:
        vals.append ((C,S))
vals.sort ()
plt.plot ([v[0] for v in vals], [v[1] for v in vals])
plt.xlabel('Mean degree C')
plt.ylabel('Size of giant component S')

plt.show ()

repeatNum = 10

vals = []

for p in np.linspace (0, 0.1, 100):
    MLC = 0.0
    for r in range (repeatNum):
        G = nx.erdos_renyi_graph (n, p)
        MLC += nx.average_clustering (G) / repeatNum
    vals.append ((p,MLC))
plt.plot ([v[0] for v in vals], [v[1] for v in vals])
plt.xlabel('p')
plt.ylabel('Mean local clustering')
plt.show ()
```

```python
#ex 3

from scipy.stats.stats import pearsonr   
import random

def pageRankVsDegree (G):
    degs = nx.degree (G)
    pagerank = nx.pagerank (G)
    fixed_order_degs = []
    fixed_order_pagerank = []
    for node in G.nodes ():
        fixed_order_degs.append (degs[node])
        fixed_order_pagerank.append (pagerank[node])
    return pearsonr(fixed_order_degs, fixed_order_pagerank)

def genRandomNet (degrees):
    G = nx.Graph ()
    n = len (degrees)
    s = sum (degrees)
    if s % 2 == 1:
        print('odd sum...')
        return 
    sl = 0
    me = 0
    while s > 0:
        v = random.randint (0, s - 1)
        v1 = 0
        while degrees[v1] < v:
            v -= degrees[v1]
            v1 += 1
        s -= 1
        degrees[v1] -= 1
        v = random.randint (0, s - 1)
        v2 = 0
        while degrees[v2] < v:
            v -= degrees[v2]
            v2 += 1
        s -= 1
        degrees[v2] -= 1
        
        if v1 == v2:
            sl += 1
        elif (v1 in G) and (v2 in G[v1]):
            me += 1
        else:
            G.add_edge (v1, v2)
    return G, sl, me

originalG = nx.karate_club_graph ()
G, sl, me = genRandomNet (nx.degree (originalG).values ())
print 'size: ', len (G.nodes ()), 'self loups: ', sl, ' multiple edges: ', me, ' sum of degrees: ', sum (nx.degree(G).values()) + 2 * sl + 2 * me, 'pagerank - degree correlation: ', pageRankVsDegree (G), ' original graph correlation: ', pageRankVsDegree (originalG) 
nx.draw (G)
plt.show ()

originalG = nx.erdos_renyi_graph (300, 0.01)
G, sl, me = genRandomNet (nx.degree (originalG).values ())
print 'size: ', len (G.nodes ()), 'self loups: ', sl, ' multiple edges: ', me, ' sum of degrees: ', sum (nx.degree(G).values()) + 2 * sl + 2 * me, 'pagerank - degree correlation: ', pageRankVsDegree (G), ' original graph correlation: ', pageRankVsDegree (originalG)
nx.draw (G)
plt.show ()
        
        
```

```python
#ex 4
C = 0.75
N = 100

G = nx.DiGraph ()
G.add_edge (0,0)
for i in range (1, N):
    in_degs = G.in_degree ()
    total = sum (in_degs.values()) + len (G.nodes ()) * C
    rv = random.random () * total
    for node in G.nodes ():
        if rv < (in_degs[node] + C):
            G.add_edge (i, node)
            break
        rv -= (in_degs[node] + C)
nx.draw (G)
plt.show ()

plt.hist (G.in_degree().values())
plt.show ()
```
